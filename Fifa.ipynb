# Data Cleaning & Preprocessing

```python
def Wrangle(df):
    # Identifying and handling missing values with respect to 'lw', 'rw', 'ls', and 'rs'
    df.dropna(subset=['rw', 'lw', 'rs', 'ls'], inplace=True)

    # Ensure relevant columns are string type before applying string operations
    for col in ['rw', 'lw', 'rs', 'ls']:
        df[col] = df[col].astype(str)

    # Right-wing position type casting to int..
    df['rw1'] = df['rw'].str.split('+', expand=True)[0].astype(int)
    df['rw2'] = df['rw'].str.split('+', expand=True)[1].astype(int)
    df['rw3'] = df['rw1'] + df['rw2']

    df.drop(columns = ['rw', 'rw1','rw2'], inplace = True)
    df.rename(columns = {'rw3':'rw'}, inplace = True)

    # Left-wing position type casting to int..
    df['lw1'] = df['lw'].str.split('+', expand=True)[0].astype(int)
    df['lw2'] = df['lw'].str.split('+', expand=True)[1].astype(int)
    df['lw3'] = df['lw1'] + df['lw2']

    df.drop(columns = ['lw', 'lw1','lw2'], inplace = True)
    df.rename(columns = {'lw3':'lw'}, inplace = True)

    # Right Strike type casting to int.
    df['rs1'] = df['rs'].str.split('+', expand=True)[0].astype(int)
    df['rs2'] = df['rs'].str.split('+', expand=True)[1].astype(int)
    df['rs3'] = df['rs1'] + df['rs2']

    df.drop(columns = ['rs', 'rs1','rs2'], inplace = True)
    df.rename(columns = {'rs3':'rs'}, inplace = True)

    # Left Strike type casting to int.
    df['ls1'] = df['ls'].str.split('+', expand=True)[0].astype(int)
    df['ls2'] = df['ls'].str.split('+', expand=True)[1].astype(int)
    df['ls3'] = df['ls1'] + df['ls2']

    df.drop(columns = ['ls', 'ls1','ls2'], inplace = True)
    df.rename(columns = {'ls3':'ls'}, inplace = True)

    return df
```

```python
# Reloading the dataframe.
df = pd.read_csv('players_20.csv', on_bad_lines='skip')
```

```python
# Now apply the Wrangle function.
Wrangle(df)
```

## **Task 2:-** Explore football skills and cluster football players based on their attributes.


```python
# KMeans clustering.

X = df[['skill_dribbling',	'skill_curve',	'skill_fk_accuracy',	'skill_long_passing']]

# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Training the K-Means model on the dataset
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

```python
# Visualising the clusters.
plt.scatter(X[y_kmeans == 0]['skill_dribbling'], X[y_kmeans == 0]['skill_curve'], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1]['skill_dribbling'], X[y_kmeans == 1]['skill_curve'], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2]['skill_dribbling'], X[y_kmeans == 2]['skill_curve'], s = 100, c = 'green', label = 'Cluster 3')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of Players based on Dribbling and Curve Skills')
plt.xlabel('Skill Dribbling')
plt.ylabel('Skill Curve')
plt.legend()
plt.show()

# Get the counts for each cluster.
cluster_counts = pd.Series(y_kmeans).value_counts().sort_index()

# Print the counts for clusters 1, 2, and 3 (which correspond to indices 0, 1, and 2).
print(f"Number of data points in Cluster 1: {cluster_counts.get(0, 0)}")
print(f"Number of data points in Cluster 2: {cluster_counts.get(1, 0)}")
print(f"Number of data points in Cluster 3: {cluster_counts.get(2, 0)}")
```

**Explanation:** Majority of players falls under the cluster 3 which have the highest skill curve and skill dribbling rating range between 60 to 80. On the second position is cluster 1, which possesses players with high performance skills upto 100. The cluster 2 has least amount of players with skills performance range from 10 to 40.

● Which type of offensive players tends to get paid the most: the striker, the right-winger, or the left-winger?

```python
# right-winger.
max_rw_wage = df.loc[df['rw'] == df['rw'].max(), 'wage_eur'].max()
```

```python
# left-winger.
max_lw_wage = df.loc[df['lw'] == df['lw'].max(), 'wage_eur'].max()
```

```python
max_lw_wage
```

```python
# right-striker.
max_rs_wage = df.loc[df['rs'] == df['rs'].max(), 'wage_eur'].max()
```

```python
max_rs_wage
```

```python
# left-striker.
max_ls_wage = df.loc[df['ls'] == df['ls'].max(), 'wage_eur'].max()
```

```python
data = {
    'Position': ['Left Winger', 'Right Winger', 'Left Striker', 'Right Striker'],
    'Max_Wage_EUR': [
        max_lw_wage,
        max_rw_wage,
        max_ls_wage,
        max_rs_wage
    ]
}
wage_comparison_df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.barplot(x='Position', y='Max_Wage_EUR', data=wage_comparison_df, palette='viridis')
plt.title('Maximum Wages for Different Offensive Player Positions')
plt.xlabel('Offensive Position')
plt.ylabel('Maximum Wage (EUR)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

 **Explanation:** In the above figure we have 4 positional skills of the players. The let-winger, and right-winger have the salary greater than 500,000 Euros, whereas, left-striker, and right-striker posses wage slightly greater than 400,000 euros.
